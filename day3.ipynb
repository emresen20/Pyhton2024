Broadcasting
Broadcasting nedir?

Broadcasting, NumPy'nin verileri kopyalamadan birden çok veri boyutunda işlem gerçekleştiren bir özelliğidir. Bu, zamandan ve yerden tasarruf sağlar. Örneğin, 3x3'lük bir diziniz (A) varsa ve 1x3'lük bir dizi (B) eklemek istiyorsanız, NumPy, (B)'nin satırını (A)'nın her satırına ekleyecektir.
Broadcasting Kuralları

İki dizinin boyut sayısı farklıysa, daha az boyuta sahip olanın şekli, öndeki (sol) taraftakilerle doldurulur.
İki dizinin şekli herhangi bir boyutta eşleşmiyorsa, o boyutta şekli 1'e eşit olan dizi diğer şekle uyacak şekilde uzatılır.
Herhangi bir boyutta boyutlar uyuşmuyorsa ve hiçbiri 1'e eşit değilse bir hata ortaya çıkar.
Broadcast kuralı: Broadcast yapabilmek için, bir işlemdeki her iki dizi için takip eden eksenlerin boyutu ya aynı boyutta olmalı ya da bunlardan biri tek olmalıdır.
